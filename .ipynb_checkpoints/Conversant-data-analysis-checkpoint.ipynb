# Conversant Data Analysis

OVERVIEW:

Looking at the data.txt file, I tried to make sense of the column titles. 
    1. Type: obtained more clarification on rtb.requests (= real time bidding requests)
    2. Time: figured out that values in the Time column areepoch times (= UNIX timestamp which represents number of seconds that have elapsed since 00:00:00 Thursday, 1 January 1970)
    3. Value: Need more clarity on what this actually is, but my assumption would be the value to win the bid at the 'Time' and served by the 'Data center'
    4. Data center: the data center serving the real time bidding request. At a glance there are 3 data centers, they could possibly span 3 different time zones.
    
ASSUMPTIONS:
    1. Negative values in the "Value" column are anamolies
    2. The epoch values of "Time" are arranged in ascending order per data center
    3. The epoch values are constant time intervals, i.e., df['Value'].iloc[1] - df['Value'].iloc[0] = df['Value'].iloc[2] - df['Value'].iloc[1]
 
ANALYSIS:

The data spans over 3 days: starting at Jun 24, 14:06 to Jun 26 3:43 for 3 data centers in total.

My first attempt was to detect if there were any anamolies in the data. I found that there were only 22 negative values in the Value column out of a total of 1678 rows. Out of which Data center S (DC_S) seemed to have 19 of them. 

Next, I found the first time stamp, last time stamp and number of requests for every data center. DC_I and DC_S start end at the same time but have different number of requests, this implies that the requests are not sampled at the same time interval for the data centers. 
In fact, given that we filtered out only 19 enteries of the total data center S enteries, number of requests for DC_S is not twice the number of requests for DC_I hints at unequal time interval. Extending this to DC_A proves my third assumption wrong. 
Further, the 19 negative values for requests being served by DC_S might be related to data being sampled more frequently than the other data centers. 

By looking at the plots, we see that DC_I serves more traffic than the others, as the values (=Y axis) are significantly higher, the mean being 31343. While the mean value of DC_S is 8457 and DC_A is 3592. Dc_A serves the least among the three. 
The patterns for DC_I and DC_S are similar, however, the values of DC_A don't deviate as much as the other, it is fairly constant with irregularities. 
On Jun 24, both DC_I and DC_S start serving lesser traffic at their respective value levels around 16:00 and it starts going up again around 22:00 ~ 23:00. On Jun 25, it is fairly high up till 14:00 and it starts to dip again. It picks up on Jun 25 ~22:00 and keeps rising through Jun 26, ~14:00. This pattern seems to be cyclic. 

INFORMATION NEEDED:

It would have been better if I understood what "Value" represented. Was it winning bid amount or value of resource used to serve the bid arriving at the given time in a particular data center. 


```python
import pandas as pd
```

```python
#reading the data into a DataFrame
df = pd.read_csv('data.txt', sep="\t")
#get the initial count
df.count()
```

```python
#Filter the data to detect anamolies
df_anamolies = df[df['Value'] < 0]
print(df_anamolies)
print("\ntotal rtb.requests with a negative value:","\n",df_anamolies.count())
#As we see there are only 22 rtb.requests with a negative value from 1678 rtb.requests
#Out of 22, 19 are in Data center S
```

```python
#Filter the data to remove anamolies
df = df[df['Value'] > 0]
df.count()
```

```python
#Convert Epoch time to UTC format as Data centers could span multiple Time Zones
import time
def parse_timestamp(epoch_time):
        return{
            "date": time.strftime("%b %d", time.gmtime(epoch_time)),
            "hour": time.strftime("%H:%M", time.gmtime(epoch_time))
                                  }
df['Date'] = df.apply (lambda row: parse_timestamp(row['Time'])["date"], axis=1)
df['Hour'] = df.apply (lambda row: parse_timestamp(row['Time'])["hour"], axis=1)
```

```python
#Find how many data centers
data_centers = df['Data center'].unique()
print(data_centers)
```

```python
#Sorting data per data center
df_I = df[df['Data center']=='dc=I']
df_S = df[df['Data center']=='dc=S']
df_A = df[df['Data center']=='dc=A']
#Find number of enteries 
print('First epoch time:',parse_timestamp(df_I['Time'].iloc[0]), '\n# of requests dc=I:',df_I['Time'].count(),'\nLast epoch time',parse_timestamp(df_I['Time'].iloc[-1]))
print('\nFirst epoch time:',parse_timestamp(df_S['Time'].iloc[0]), '\n# of requests dc=S:',df_S['Time'].count(),'\nLast epoch time',parse_timestamp(df_S['Time'].iloc[-1]))
print('\nFirst epoch time:',parse_timestamp(df_A['Time'].iloc[0]), '\n# of requests dc=A:',df_A['Time'].count(),'\nLast epoch time',parse_timestamp(df_A['Time'].iloc[-1]))
```

```python
#Getting Maximun, Minimum and Mean of Values for every data center
print("DC_I-   Max: {} Min: {} Mean: {}".format(df_I['Value'].max(), df_I['Value'].min(), df_I['Value'].mean()))
print("DC_S-   Max: {} Min: {} Mean: {}".format(df_S['Value'].max(), df_S['Value'].min(), df_S['Value'].mean()))
print("DC_A-   Max: {} Min: {} Mean: {}".format(df_A['Value'].max(), df_A['Value'].min(), df_A['Value'].mean()))
```

```python
#Plotting data per day for all Data center for comparison
days = ["Jun 24", "Jun 25", "Jun 26"]
for day in days:
    df_I_day = df_I[df_I['Date'] == day]
    df_S_day = df_S[df_S['Date'] == day]
    df_A_day = df_A[df_A['Date'] == day]

    ax_I=df_I_day.plot(kind='scatter', x='Time', y='Value', color='red',legend='great')
    ax_S=df_S_day.plot(ax=ax_I,kind='scatter', x='Time', y='Value',color='green',)
    ax_A=df_A_day.plot(ax=ax_S,kind='scatter', x='Time', y='Value',color='yellow',)
    ax_A.set(xlabel="Time", ylabel="Values", title=day)
    ax_A.legend(['DC_I', 'DC_S','DC_A'])
```

```python
#Detailed graphs for further analysis
#Plotting data for every data center on a given day 
data_centers = [df_I, df_S, df_A]
day = input("Enter the day as: Jun 24 / Jun 25 / Jun 26: ")
def display_by_day(data_centers, day):
    data_center_names=['I','S','A']
    i = 0
    for data_center in data_centers:
        df_day = data_center[data_center['Date'] == day]
        ax = df_day.plot(kind='line', x='Hour', y='Value')
        ax.set(xlabel="Time", ylabel="Values", title="Data center {} on {}".format(data_center_names[i],day))
        i = i + 1
display_by_day(data_centers, day)
```

```python
#Plotting data for all three days for a given data center
def user_input(data_center_name):
    if data_center_name is "I":
        return df_I
    elif data_center_name is "S":
        return df_S
    elif data_center_name is "A":
        return df_A
    else:
        print("Please chose a valid data center name: I, S or A")
               
def display_by_data_center():
    data_center_name = input("I, S or A: ")
    days = ["Jun 24", "Jun 25", "Jun 26"]
    data_center = user_input(data_center_name)
    for day in days:
        df_day = data_center[data_center['Date'] == day]
        ax = df_day.plot(kind='line', x='Hour', y='Value', color='yellow')
        ax.set(xlabel="Time", ylabel="Values", title="Data center {} on {}".format(data_center_name,day))
               
display_by_data_center()
```
